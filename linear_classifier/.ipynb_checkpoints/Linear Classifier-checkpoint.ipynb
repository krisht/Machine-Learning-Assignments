```python
import numpy as np
import scipy
import matplotlib.pyplot as plt
import pylab

N = 1000
mu1 = [1, 1]
mu2 = [-1, -1]
cov = np.eye(2)/2

def sigmoid(x):
    return scipy.special.expit(x)
```

```python
# Generate Data

#Generate first set
x1, y1 = np.random.multivariate_normal(mu1, cov, N).T
labels = np.zeros(len(x1))
samples1 = np.asarray([x1, y1, labels])

#Generate second set
x2, y2 = np.random.multivariate_normal(mu2, cov, N).T
labels = np.ones(len(x2))
samples2 = np.asarray([x2, y2, labels])


total = np.concatenate((samples1, samples2), axis = 1).T

np.random.shuffle(total)

train = total[0:800]
test = total[800:1000]

print("Training Data Size: %d" % train.shape[0])
print("Testing Data Size: %d" % test.shape[0])
```
